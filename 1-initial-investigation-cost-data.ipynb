#### Initial cleaning

With our basic dataframe imported, we can see that the '--' entries in the "bonus" column are causing that column to be considered as strings instead of floats.  We'll begin by replacing the '--' value with '0', then converting that column data type to floats.

```python
costs.bonus.replace(to_replace='--', value=0, inplace=True)
```

```python
costs.head(10)
```

```python
costs.bonus = costs.bonus.astype(dtype='float64')
```

#### Checking for Outliers

```python
costs['total_federal_cost'].plot.box()
```

Our simple box plot here shows that the data for Total Federal Cost is right-skewed, which is expected, as our line graphs indicate a relatively steady increase over time.

```python
ax = costs.plot.box(subplots=True, layout=(2, 4), figsize=(16, 10))
```

We have a few outliers, all on the high side, and we might want to take a look at these, but it is likely that they are all naturally-occuring results that have to do with the shape of our data, and the steady increase of costs over time.  It is unlikely that any of them need to be removed from our later analysis.  In the interest of due diligence, though, we'll use the `zscore` function from scipy.stats to verify that any zscores above 2 are those that appear where we would expect (most recent for `breakfast_payments`, `entitlement`, and `total_commodity_costs`, mid-1980s for `bonus`.

```python
from scipy.stats import zscore
```

```python
def get_z_score(column_name, df):
    df['z_score'] = zscore(df[column_name])
    outliers = df.query('z_score > 2')
    return outliers
```

```python
outliers = get_z_score('breakfast_payments', costs)
outliers
```

```python
outliers = get_z_score('entitlement', costs)
outliers
```

```python
outliers = get_z_score('bonus', costs)
outliers
```

```python
outliers = get_z_score('total_commodity_costs', costs)
outliers
```

These results are all as expected, so we will choose to keep our outlier data in the dataset, and won't be altering or dropping any of these values.  The last thing we'll do is drop the z_score column, as it is no longer relevant.

```python
costs.drop('z_score', axis=1, inplace=True)
```
